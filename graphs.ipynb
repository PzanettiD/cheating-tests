```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_pos = ['A', 'B', 'C', 'D', 'E']
all_values = [401, 417, 418, 469, 386]
values0 = [139, 166, 148, 168, 141]
values1 = [141, 143, 158, 149, 132]
values2 = [121, 108, 112, 152, 113]
temp_x = np.arange(len(x_pos))
temp_x
```

```python
plt.xticks(temp_x, x_pos)
plt.bar(temp_x-0.2, all_values, width=0.2, label='Total')
plt.bar(temp_x, values0, width=0.2, label='2016')
plt.bar(temp_x+0.2, values1, width=0.2, label='2017')
plt.bar(temp_x+0.4, values2, width=0.2, label='2018')
plt.title('Testes por Letra, 2016/17/18')
plt.ylabel('Número de Testes')
plt.legend()
plt.savefig('graph.jpg')
```

```python
a_values = [values0[0], values1[0], values2[0]]
b_values = [values0[1], values1[1], values2[1]]
c_values = [values0[2], values1[2], values2[2]]
d_values = [values0[3], values1[3], values2[3]]
e_values = [values0[4], values1[4], values2[4]]
plt.plot(a_values, label='A')
plt.plot(b_values, label='B')
plt.plot(c_values, label='C')
plt.plot(d_values, label='D')
plt.plot(e_values, label='E')
plt.legend()
plt.title("Respostas dos testes ao passar dos anos")
plt.ylabel("Número de Testes")
plt.xlabel("Tempo apartir de 2016 (em anos)")
plt.savefig('graph1.jpg')
```
